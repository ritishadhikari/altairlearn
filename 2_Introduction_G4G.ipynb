```python
import numpy as np
import pandas as pd
import altair as alt
from vega_datasets import data
```

#### Simple Bar Chart

```python
score_data = pd.DataFrame({ 
    'Website': ['StackOverflow', 'FreeCodeCamp', 
                'GeeksForGeeks', 'MDN', 'CodeAcademy'], 
    'Score': [65, 50, 99, 75, 33] 
})   

score_data
```

```python
alt.Chart(data=score_data).\
    mark_bar().\
    encode(
        x="Website",
        y="Score"
    )
```

#### ScatterPlot

```python
iris=data.iris()
```

```python
iris.head()
```

```python
alt.Chart(data=iris).\
    mark_point().\
    encode(
        x="sepalLength",
        y="petalLength",
        shape="species",
    )

```

#### Grouped Bar Chart

```python
data = pd.DataFrame([[264, 'Rohit', 'ODI'],  
                     [183, 'Virat', 'ODI'],  
                     [118, 'Rohit', 'T20'],  
                     [94, 'Virat', 'T20'], 
                     [212, 'Rohit','Test'], 
                     [254, 'Virat','Test']], 
                     columns=['Highest Score', 'Player', 'Format']) 
```

```python
data
```

```python
alt.\
Chart(data=data).\
mark_bar().\
encode(
    column="Format",
    x="Player",
    y=alt.Y(shorthand="Highest Score",axis=alt.Axis(grid=False)),
    color="Player",
)
```
