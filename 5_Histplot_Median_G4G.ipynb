```python
import numpy as np
import pandas as pd
import altair as alt
from vega_datasets import data
from pathlib import Path
import seaborn as sns
```

```python
np.random.seed(seed=42)
df=pd.DataFrame(
    data={
        'height':np.random.normal(loc=150,scale=10,size=1000)
    }
)
```

```python
df.head()
```

```python
base=alt.Chart(data=df)
```

```python
hist=base\
.mark_bar(color='green')\
.encode(
    x=alt.X(
        shorthand="height",
        bin=alt.BinParams(maxbins=20),
        title="Spread"
    ),
    y=alt.Y(shorthand="count()"),
    
    
)

hist.show()
```

```python
median=base\
.mark_rule(color='black')\
.encode(
    x=alt.X(
        shorthand="mean(height):Q",
    ),
    size=alt.value(value=5)   
)

median.show()
```

```python
hist+median
```

```python
(hist+median).save(
    fp=Path("plotImages").joinpath('histmedian3.png'),
    format="png"
)
```

#### Density Plots

```python
density=base\
.transform_density(
    density="height",
    as_=['randomNumber','Density']
)\
.mark_area(
    color="green"
)\
.encode(
    x=alt.X(shorthand="randomNumber:Q"),
    y=alt.Y(shorthand="Density:Q")
)
density.show()
```

#### Scatter Plot with Colors

```python
cars=data.cars()
```

```python
cars.head()
```

```python
cars['Origin'].unique()
```

```python
cols=['#4B0082','yellow','#FFC0CB']
```

```python
sp=alt\
.Chart(
    data=cars,
    width=800,
    height=700
    )\
.mark_point()\
.encode(
    x=alt.X(shorthand="Miles_per_Gallon"),
    y=alt.Y(shorthand="Horsepower"),
    color=alt.Color(
        shorthand="Origin",
        scale=alt.Scale(
            domain=cars['Origin'].unique(),
            range=cols
        )
    )
)
sp.show()
```

```python
sp.save(
    fp=Path('plotImages').joinpath('scatter4.png'),
    format='png'
)
```

```python
sp=alt\
.Chart(
    data=cars,
    width=800,
    height=700
    )\
.mark_point()\
.encode(
    x=alt.X(shorthand="Miles_per_Gallon"),
    y=alt.Y(shorthand="Horsepower"),
    color=alt.Color(
        shorthand="Origin",
        scale=alt.Scale(
            scheme="accent"
        )
    )
)
sp.show()
```

```python
sp1=alt\
.Chart(
    data=cars,
    width=800,
    height=700
    )\
.mark_point()\
.encode(
    x=alt.X(shorthand="Miles_per_Gallon"),
    y=alt.Y(shorthand="Horsepower"),
    color=alt.Color(
        shorthand="Origin",
        scale=alt.Scale(
            scheme="accent"
        )
    ),
    tooltip=[
        alt.Tooltip(shorthand="Miles_per_Gallon"),
        alt.Tooltip(shorthand="Horsepower")
    ]
)
```

```python
sp1.show()
```
